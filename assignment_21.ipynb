## Q7. Describe the math package's general categories of functions.



>Basic Arithmetic Functions:
These functions handle fundamental arithmetic operations like addition, subtraction, multiplication, division, and exponentiation. They are the building blocks for more complex calculations.

>Trigonometric Functions:
Trigonometric functions deal with angles and relationships within triangles. They include functions like sine, cosine, tangent, cosecant, secant, and cotangent, along with their inverse functions (arcsine, arccosine, arctangent, etc.).

>Exponential and Logarithmic Functions:
Exponential functions involve calculations related to powers of constants like the base of natural logarithms (e.g., e^x). Logarithmic functions, on the other hand, help with solving equations involving exponents (e.g., log(x)).

>Hyperbolic Functions:
Hyperbolic functions are analogs to trigonometric functions but apply to hyperbolas. They include functions like sinh (hyperbolic sine), cosh (hyperbolic cosine), and tanh (hyperbolic tangent).

>Statistical Functions:
These functions are used for statistical calculations, such as mean, median, standard deviation, variance, correlation, and probability distributions. They are crucial for data analysis and modeling.

>Complex Number Functions:
Complex number functions handle operations involving complex numbers, which have both real and imaginary components. These functions include complex number arithmetic, conjugate calculations, and polar/rectangular conversions.

## Q8. What is the relationship between exponentiation and logarithms?



>Exponentiation:

Exponentiation involves raising a base number to a certain power, resulting in a new number called the exponentiation or power of the base. In the expression a^b, "a" is the base and "b" is the exponent. The result of the operation is a raised to the power of b.

>Logarithms:

Logarithms are the inverse of exponentiation. A logarithm tells you what exponent you need to raise a particular base to in order to get a given number. In the expression log_a(b), "a" is the base of the logarithm, and "b" is the number whose logarithm is being taken. The result of the operation is the exponent you need to raise "a" to in order to get "b".

>If a^b = c, then log_a(c) = b.

## Q9. What are the three logarithmic functions that Python supports?

The following are the variants of the basic log function in Python: 
>log2(x) 

>log(x, Base) 

>log10(x)

```python
import math

x = 10
natural_log = math.log(x)   #first type
print("Natural logarithm of", x, ":", natural_log)

x = 100
common_log = math.log10(x) #second type
print("Common logarithm of", x, ":", common_log)

x = 8
custom_base = 2
custom_log = math.log(x, custom_base) #thrid type
print("Logarithm of", x, "with base", custom_base, ":", custom_log)
```
