## Q3. What is a histogram, exactly? Name a numpy method for creating such a graph.



> histogram is a grpahical representation to see how the data is distributed

>the data is divided into bins or intervals, and the height of each bar in the histogram represents the frequency or count of data points falling within that bin.

>numpy.histogram method is used to get it done by numpy

```python
import numpy as np
import matplotlib.pyplot as plt

#creating a random dataset using numpy
data = np.random.randn(500)
print(data)
hist_values, bin_edges = np.histogram(data, bins=10)
plt.hist(data, bins=10, alpha=0.5, color='red', edgecolor='black')

plt.xlabel('value')
plt.ylabel('Frequency')
plt.title('Histogram ')

# Show the plot
plt.show()
```

## Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot product, outer product, and regular multiplication of two numpy arrays.



```python
import numpy as np
import numpy_financial as npf

array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
result = array1 * array2 
print("Regular multiplication  \n================================\n",result)

array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
result = np.outer(array1, array2)
print("\n\nOuter product using numpy\n================================\n",result)

array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])
result = np.dot(array1, array2)
print("\n\nDot product using numpy  \n================================\n",result)


```

## Q6. Before you buy a home, which numpy function will you use to measure your monthly mortgage payment?



>import numpy-financial as npf

>npf.pmt()

```python
import numpy as np
import numpy_financial as npf
#loan amount
loan_amount = 3000000  

#interest
annual_interest_rate = 0.035

#loan term in years
loan_term_years = 5  

num_monthly_payments = loan_term_years * 12

monthly_interest_rate = annual_interest_rate / 12 
monthly_payment = npf.pmt(monthly_interest_rate, num_monthly_payments, -loan_amount)


print(f"Your monthly mortgage payment is approximately: {monthly_payment:.2f} INR")
```
